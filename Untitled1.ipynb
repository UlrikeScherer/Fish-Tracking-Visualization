```python
import numpy as np
from sklearn.manifold import TSNE
from src.utile import *
from src.feeding import *
%load_ext autoreload
%autoreload 2
%load_ext line_profiler

def transfrom_to_traces(batch, trace_size):
    setX = batch[["xpx", "ypx"]].to_numpy()
    setX = setX[1:]-setX[:-1]
    lenX = setX.shape[0]
    sizeSet = int(lenX/trace_size)
    newSet = np.zeros((sizeSet,trace_size,2))
    for i in range(sizeSet-1):
        s = setX[i*trace_size]
        alph = np.arctan2(*s)
        newSet[i,:,:] = np.dot(setX[i*trace_size:(i+1)*trace_size],rotation(-alph))
    X = np.reshape(newSet, (sizeSet, trace_size*2))
    return X, newSet
```

```python
batch1 = csv_of_the_day(get_camera_names()[0], get_days_in_order()[23])[9]
batch2 = csv_of_the_day(get_camera_names()[0], get_days_in_order()[23])[1]

X, nS = transfrom_to_traces(batch1, trace_size=20)
```

```python

X2, nS2 = transfrom_to_traces(batch2, 20)
X_embedded = TSNE(n_components=2, init='random', learning_rate=100).fit_transform(np.concatenate((X, X2)))
```

```python
half = int(X_embedded.shape[0]/2)
plt.scatter(X_embedded[:half,0], X_embedded[:half, 1])
plt.scatter(X_embedded[half:,0], X_embedded[half:, 1])
```
